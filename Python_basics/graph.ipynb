### 제목 넣기

```python
plt.title('plotting')
plt.plot([10,20,30,40])
plt.show()
```

### 범례 넣기

```python
plt.title('example')
plt.plot([10,20,30,40], label = 'asc')
plt.plot([40,30,20,10], label = 'desc')
plt.legend()
plt.show()
```

### 그래프 색상 바꾸기

```python
plt.title('color')
plt.plot([10,20,30,40], color = 'green', label = 'green')
plt.plot([40,30,20,10], 'purple', label = 'bora')
plt.legend()
plt.show()
```

### 그래프 선 모양 바꾸기

```python
plt.title('linestyle') # 제목 설정
# 빨간색 dashed 그래프
plt.plot([10,20,30,40], color='r', linestyle='--', label = 'dashed')
# 초록색 dotted 그래프
plt.plot([40,30,20,10], color='g', ls=':', label='dotted')
plt.legend()
plt.show()
```

### 마커 모양 바꾸기

```python
plt.title('marker')
plt.plot([10,20,30,40], 'r.', label='circle') # 빨간색 원형 마커 그래프
# 초록색 삼각형 마커 그래프
plt.plot([40,30,20,10], 'g^', label = 'triangle up')
plt.legend()
plt.show()
```
